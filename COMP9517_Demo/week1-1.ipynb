# Image Resizing

```python
opera = cv2.imread('OperaHouse.jpg')
opera_rgb = cv2.cvtColor(opera, cv2.COLOR_BGR2RGB)
plt.imshow(opera_rgb)
plt.show()
```

```python
resized_opera = cv2.resize(opera_rgb, (img.shape[1], img.shape[0]), cv2.INTER_AREA) # resize for image addition below
plt.imshow(resized_opera)
plt.show()
```

# Image Addition using OpenCV

```python
dst = cv2.addWeighted(resized_opera,0.2,rgb_img,0.8,0)
plt.imshow(dst)
plt.show()
```

```python
dst = cv2.add(resized_opera, rgb_img)
plt.imshow(dst)
plt.show()
```

# Image Addition using numpy

```python
dst = resized_opera + rgb_img
```

```python
plt.imshow(dst)
```

```python
# Note the difference between cv2.add and numpy addition

x = np.uint8([250])
y = np.uint8([10])
print( cv2.add(x,y) ) # 250+10 = 260 => 255

print( x+y )          # 250+10 = 260 % 256 = 4
```

# Visualising individual RGB colour channels

```python
img = cv2.imread("lorikeet.jpeg") # load the image
plt.axis('off') # hide the axis
plt.imshow(img)
plt.show()
```

```python
im_RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # convert BGR  RGB
plt.imshow(im_RGB)
plt.show()
```

```python
R = im_RGB.copy() # copy the image to another variable
R[:,:,(1,2)] = 0 # zero out the second and third channels
plt.imshow(R)
plt.show()
```

```python
G = im_RGB.copy() # copy the image to another variable
G[:,:,(0,2)] = 0 # zero out the first and third channels
plt.imshow(G)
plt.show()
```

```python
B = im_RGB.copy()  # copy the image to another variable
B[:,:,(0,1)] = 0 # zero out the first and second channels
plt.imshow(B)
plt.show()
```

# Visualising individual HSV colour channels

```python
img_HSV = cv2.cvtColor(img, cv2.COLOR_BGR2HSV) # convert BGR  HSV
```

```python
H = img_HSV.copy()
H[:, :, (1,2)] = 255 # set S and V channels to max (255)
H_RGB =  cv2.cvtColor(H,cv2.COLOR_HSV2RGB) # convert back to RGB
plt.show()
plt.imshow(H_RGB)
```

```python
S = img_HSV.copy()
S[:, :, 0] = 179 # set H to max (179)
S[:, :, 2] = 255 # set V to max (255)
S_RGB =  cv2.cvtColor(S,cv2.COLOR_HSV2RGB) # convert back to RGB
plt.show()
plt.imshow(S_RGB)
```

```python
V = img_HSV.copy()
V[:, :, 0] = 179 # set H to max (179)
V[:, :, 1] = 0 # set S to 0
V_RGB =  cv2.cvtColor(V,cv2.COLOR_HSV2RGB) # convert back to RGB 
plt.show()
plt.imshow(V_RGB)
```
